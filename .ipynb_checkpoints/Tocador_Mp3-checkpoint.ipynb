# <font color='green'>***Tocador Simples MP3***</font>

```python
# Importando as bibliotecas necessárias
from pygame import mixer
from time import sleep
import os

# Inicializando o mixer Pygame
mixer.init()

# Lista de músicas para o programa
lista_musica = ['Sharks.mp3', 'thunder.mp3', 'dalila.mp3', '10-metros-quadrados.mp3',
                'casamentodegrau.mp3', 'demais.mp3']

# Índice da próxima música a ser tocada
proxima_musica = 0

# Carregando a primeira música da lista
mixer.music.load(lista_musica[proxima_musica])

# Menu de opções para o usuário
menu = '''
======================
=   Menu de Opções   =
======================
[1] Tocar Música     =
[2] Pausar Música    =
[3] Retomar Música   =
[4] Próxima Música   =
[5] Música Anterior  =
[6] Voltar ao Menu   =
[7] Reiniciar        =
======================
'''

# Loop infinito para o player de MP3
while True:
    # Solicitando ao usuário se ele deseja mostrar o menu ou encerrar o programa
    mostra_menu = input('Deseja mostrar o menu [S/N] ou encerrar o programa [E]: ').upper()[0]

    # Se o usuário não deseja mostrar o menu, o programa continua em execução
    if mostra_menu == "N":
        continue

    # Se o usuário deseja mostrar o menu, o programa exibe o menu e solicita uma opção
    elif mostra_menu == "S":
        print(menu)
        contador = 1
        while True:
            if contador > 2:
                os.system('clear')

            opcao = input('Digite a opção desejada [6] Mostrar o Menu: ')

            try:
                opcao = int(opcao)

                # Tocar música
                if opcao == 1:
                    print('Tocando a Música...')
                    mixer.music.play()

                # Pausar música
                elif opcao == 2:
                    print('Música Pausada.')
                    mixer.music.pause()

                # Retomar música
                elif opcao == 3:
                    print('Música Retomada.')
                    mixer.music.unpause()

                # Próxima música
                elif opcao == 4:
                    print('Próxima Música...')
                    mixer.music.stop()
                    proxima_musica = (proxima_musica + 1) % len(lista_musica)
                    mixer.music.load(lista_musica[proxima_musica])
                    mixer.music.play()

                # Música anterior
                elif opcao == 5:
                    print('Música Anterior...')
                    mixer.music.stop()
                    proxima_musica = (proxima_musica - 1) % len(lista_musica)
                    mixer.music.load(lista_musica[proxima_musica])
                    mixer.music.play()

                # Mostrar o menu
                elif opcao == 6:
                    print('Mostrando o Menu...')
                    print(menu)

                # Reiniciar
                elif opcao == 7:
                    print('Reiniciando...')
                    break

                # Opção inválida
                else:
                    print(f'Por favor, digite uma opção válida do MENU: {menu}')

            except ValueError:
                print('Por favor, digite um valor inteiro válido.')

            except KeyboardInterrupt:
                print('\nPrograma encerrado com sucesso.')
                exit()

    # Se o usuário deseja encerrar o programa
    elif mostra_menu == "E":
        print('Encerrando o Programa...')
        sleep(2)
        print('Programa Encerrado Com Sucesso.')
        break

    else:
        print('Tente novamente, você deve digitar [S/N] ou [E] para encerrar o programa.')
```

### **Explicando o programa**

        Importação das bibliotecas necessárias: O programa começa importando as bibliotecas necessárias - pygame para controle de áudio, time para controle de tempo e os para interação com o sistema operacional.

        Inicialização do mixer Pygame: O mixer Pygame é inicializado com mixer.init(). O mixer é a parte do Pygame que lida com o áudio.

        Definição da lista de músicas: Uma lista de músicas é definida. Esta lista contém os nomes dos arquivos de música que o programa pode tocar.

        Carregamento da primeira música: A primeira música da lista é carregada usando mixer.music.load().

        Definição do menu de opções: Um menu de opções é definido como uma string. Este menu é exibido para o usuário para que ele possa escolher o que fazer a seguir.

        Loop principal: O programa entra em um loop infinito. Dentro deste loop, o programa solicita ao usuário se ele deseja ver o menu de opções ou encerrar o programa.

        Se o usuário optar por não ver o menu, o programa continua em execução.
        Se o usuário optar por ver o menu, o programa exibe o menu e entra em outro loop. Dentro deste loop, o programa solicita ao usuário que digite uma opção do menu.
        Dependendo da opção escolhida pelo usuário, o programa pode tocar música, pausar música, retomar música, ir para a próxima música, voltar para a música anterior, mostrar o menu novamente ou reiniciar o programa.
        Se o usuário optar por encerrar o programa, o programa é encerrado.
